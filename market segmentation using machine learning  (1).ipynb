```python
# Outlier Detection and Removal 

# Select numeric columns only
numeric_cols = df.select_dtypes(include="number").columns

# Dictionary to store outlier counts
outlier_counts = {}

# Loop through each numeric column
for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Identify outliers
    outlier_mask = (df[col] < lower_bound) | (df[col] > upper_bound)
    outlier_counts[col] = int(outlier_mask.sum())

# Convert the result to a DataFrame and sort it
outliers_df = pd.DataFrame(list(outlier_counts.items()), columns=["Column", "Outlier Count"])
outliers_df = outliers_df.sort_values("Outlier Count", ascending=False).reset_index(drop=True)

# Display the result
print(outliers_df)
```

```python
# capping for outlier 
numeric_cols = df.select_dtypes(include="number").columns

for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_cap = q1 - 1.5 * iqr
    upper_cap = q3 + 1.5 * iqr
    
    # Winsorise: values < lower_cap become lower_cap, values > upper_cap become upper_cap
    df[col] = df[col].clip(lower_cap, upper_cap)
```

Scaling the Dataframe

```python
from sklearn.preprocessing import StandardScaler
sd=StandardScaler()
```

```python
scaled_df = sd.fit_transform(df)
```

PCA(Converting the DataFrame into 2D DataFrame for visualization)

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_df)
pca_df = pd.DataFrame(data=principal_components ,columns=["PCA1","PCA2"])
pca_df
```

Model Training and Evaluation

```python
from sklearn.cluster import KMeans
```

```python
# finding the value of k by using Elbow Method
inertia = []
range_val = range(1, 15)

for i in range_val:
    kmean = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmean.fit(scaled_df)  
    inertia.append(kmean.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range_val, inertia, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()
```

```python
kmeans_model=KMeans(4)
kmeans_model.fit_predict(scaled_df)
pca_df_kmeans= pd.concat([pca_df,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
```

```python
# Visualising the clustered dataframe
plt.figure(figsize=(10,10))
lx=sns.scatterplot(x="PCA1",y="PCA2",hue="cluster",data=pca_df_kmeans,palette=['green','red','black','yellow'])
plt.title("Clustering using K-Means Algorithm")
plt.show()
```

```python
# find all cluster centers
cluster_centers = pd.DataFrame(data=kmeans_model.cluster_centers_,columns=[df.columns])
# inverse transform the data
cluster_centers = sd.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data=cluster_centers,columns=[df.columns])
cluster_centers
```
